```python
# -----------------------------------------
# QUESTION 1 – Accident Prediction
# -----------------------------------------

# Prior probabilities
P_yes = 5/10
P_no  = 5/10

# Conditional probabilities (Accident = YES)
P_rain_yes = 1/5
P_good_yes = 1/5
P_normal_yes = 1/5
P_no_yes = 2/5

# Conditional probabilities (Accident = NO)
P_rain_no = 2/5
P_good_no = 3/5
P_normal_no = 2/5
P_no_no = 4/5

# Naive Bayes calculations
P_accident_yes_given_X = P_rain_yes * P_good_yes * P_normal_yes * P_no_yes * P_yes
P_accident_no_given_X  = P_rain_no  * P_good_no  * P_normal_no  * P_no_no  * P_no

print("QUESTION 1:")
print("P(Accident=Yes|X) =", P_accident_yes_given_X)
print("P(Accident=No|X)  =", P_accident_no_given_X)

if P_accident_no_given_X > P_accident_yes_given_X:
    print("Prediction: ACCIDENT = NO\n")
else:
    print("Prediction: ACCIDENT = YES\n")


# -----------------------------------------
# QUESTION 2 – Play Tennis
# -----------------------------------------

# Prior
P_play_yes = 9/14
P_play_no  = 5/14

# Conditional (Play = YES)
P_sunny_yes = 2/9
P_hot_yes = 2/9
P_high_yes = 3/9
P_false_yes = 6/9

# Conditional (Play = NO)
P_sunny_no = 3/5
P_hot_no = 2/5
P_high_no = 4/5
P_false_no = 2/5

# Naive Bayes
P_yes_X = P_sunny_yes * P_hot_yes * P_high_yes * P_false_yes * P_play_yes
P_no_X  = P_sunny_no  * P_hot_no  * P_high_no  * P_false_no  * P_play_no

print("QUESTION 2:")
print("P(Play=Yes|X) =", P_yes_X)
print("P(Play=No|X)  =", P_no_X)

if P_no_X > P_yes_X:
    print("Prediction: PLAY = NO\n")
else:
    print("Prediction: PLAY = YES\n")


# -----------------------------------------
# QUESTION 3 – Loan Approval
# -----------------------------------------

# Conditional (Loan Approved = YES)
P_loan_yes = 3/5
P_employed_yes = 2/3
P_good_yes_loan = 1     # 3/3
P_high_yes = 1/3

# Conditional (Loan Approved = NO)
P_loan_no = 2/5
P_employed_no = 1/2
P_bad_no = 1            # 2/2
P_low_no = 1/3

# Bayes
P_yes_loan = P_loan_yes * P_employed_yes * P_good_yes_loan * P_high_yes
P_no_loan  = P_loan_no  * P_employed_no  * P_bad_no         * P_low_no

print("QUESTION 3:")
print("P(Loan=Yes|X) =", P_yes_loan)
print("P(Loan=No|X)  =", P_no_loan)

if P_yes_loan > P_no_loan:
    print("Prediction: LOAN = APPROVED\n")
else:
    print("Prediction: LOAN = NOT APPROVED\n")


# -----------------------------------------
# QUESTION 4 – Disease Diagnosis
# -----------------------------------------

# Prior
P_positive = 3/5
P_negative = 2/5

# Conditional (Positive)
P_fever_pos = 1       # 3/3
P_cough_pos = 1/3
P_fatigue_pos = 1     # 3/3
P_travel_pos = 1/3

# Conditional (Negative)
P_fever_neg = 1       # 2/2
P_cough_neg = 0       # 0/2
P_fatigue_neg = 1     # 2/2
P_travel_neg = 1/3

# Bayes
P_disease_yes = P_fever_pos * P_cough_pos * P_fatigue_pos * P_travel_pos * P_positive
P_disease_no  = P_fever_neg * P_cough_neg * P_fatigue_neg * P_travel_neg * P_negative

print("QUESTION 4:")
print("P(Disease=YES|X) =", P_disease_yes)
print("P(Disease=NO|X)  =", P_disease_no)

if P_disease_yes > P_disease_no:
    print("Prediction: DISEASE = YES")
else:
    print("Prediction: DISEASE = NO")
```
