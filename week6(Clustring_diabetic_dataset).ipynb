```python

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Use Glucose and BMI as features
xx = df[['Glucose', 'BMI']]

# Fit KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(xx)

# Scatter plot with cluster labels
sns.scatterplot(data=xx, x="Glucose", y="BMI", hue=kmeans.labels_, palette="Set1")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            marker="X", c="black", s=100, label="Centroids")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Select features
xx = df[['Glucose', 'BMI']]

# Fit KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(xx)

# Plot
sns.scatterplot(data=xx, x="Glucose", y="BMI", hue=kmeans.labels_, palette="Set1")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            marker="X", c="r", s=80, label="centroids")
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Example diabetic data
newdata = [[85, 30, 50], [120, 25, 35], [140, 32, 60], [100, 28, 45]]
d = pd.DataFrame(newdata, columns=['Glucose', 'BMI', 'Age'])

# Fit KMeans on this data (or ideally on full training data)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(d)

# Predict clusters
predictedClusters = kmeans.predict(d)

print("Number of predicted clusters:", len(predictedClusters))
print("Data:\n", d)
print("Predicted Clusters:", predictedClusters)
```
