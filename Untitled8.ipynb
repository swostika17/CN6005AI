```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer, load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
breast_cancer = load_breast_cancer()
X_binary = breast_cancer.data
y_binary = breast_cancer.target
```

```python
X_train, X_test, y_train, y_text,=train_test_split(X_binary,y_binary, test_size=0.3, random_state=50)
```

```python
perceptron = Perceptron(max_iter=1000, random_state=42)
perceptron.fit(X_train, y_train)
```

```python
y_pred = perceptron.predict(X_test)
print(classification_report(y_text, y_pred))
print(confusion_matrix(y_text, y_pred))
```

```python
wine=load_wine()
import pandas as pd
df_wine=pd.DataFrame(wine.data, columns=wine.feature_names)
df_wine['target']=wine.target
df_wine.head()
df_wine.shape
```

```python
X=df_wine.drop('target', axis=1)
y=df_wine['target']
```

```python
X_train_wine, X_test_wine, y_train_wine, y_test_wine=train_test_split(X,y, test_size=0.3, random_state=50)
```

```python
perceptron_wine = Perceptron(max_iter=1000, random_state=42)
perceptron_wine.fit(X_train_wine, y_train_wine)
```

```python
y_pred_wine = perceptron_wine.predict(X_test_wine)
print(classification_report(y_test_wine, y_pred_wine))
print(confusion_matrix(y_test_wine, y_pred_wine))
```

```python
conf_matrix_wine = confusion_matrix(y_test_wine, y_pred_wine)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_wine, annot=True, fmt='d', cmap='Blues', xticklabels=wine.target_names, yticklabels=wine.target_names)
plt.title('Confusion Matrix - Multi-Class Classification')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
