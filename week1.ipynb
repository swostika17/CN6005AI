```python
# List Example
numbers = [1, 2, 3, 4, 5]
print("Numbers:", numbers)
numbers.append(6)
print("After adding 6:", numbers)

# Conditional Example
year = 2024
if year % 4 == 0:
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")

# For Loop Example
sum_squares = 0
for i in range(1, 6):
    sum_squares += i * i
print("Sum of squares 1 to 5:", sum_squares)

# Factorial Example
n = 5
fact = 1
for i in range(1, n + 1):
    fact *= i
print("Factorial of", n, "is", fact)

# Dictionary Example
student = {'name': 'John', 'age': 20, 'city': 'New York'}
print("Student info:", student)
```
